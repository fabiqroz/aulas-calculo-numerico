### Vamos pensar em um problema:  Encontrar o zero da função $f(x) = x-5$

**Solução Analítica**:

$𝑓(𝑥)= 𝑥-5 = 0$ que implica em $𝑥=  5$

**Solução Numérica:**

Passo 1)

1.   Vamos chutar que $x = 1$, $𝑓(1)=1−5=−4$: <font color='red'>**um valor negativo!**</font>
2.   Vamos chutar que $x = 6$, então $𝑓(6)= 6−5 = 1$: <font color='green'>**um valor positivo!**</font>
   
     É razoável que o zero dessa função esteja no intervalo $1< x < 6$.

3. Vamos pensar em um novo $x = \frac{1+6}{2} = \frac{7}{2} = 3.5$. $𝑓(3.5)=3.5−5=-1.5$: <font color='red'> **um valor negativo!**</font>
 
   É razoável pensar que o zero dessa função  esteja no intervalo $3.5 < x < 6$

4. Vamos repetir o Passo 3 para diminuir ainda mais o meu intervalo de interesse até quando eu achar que tá "bacana"

**Resumidamente:** O campo da Análise Numérica é o estudo de como "pegar" problemas matemáticos e realizá-los com eficiência e precisão em um computador.

**Obviamente:** Embora o campo da análise numérica seja bastante poderoso e abrangente, existem alguns problemas matemáticos em que a análise numérica não faz muito sentido, por exemplo

##  Influência dos Erros nas Soluções <a class="anchor" id="section_2_1"></a>



## A gerra do golfo e a falha no sistema de defesa das tropas americanas: Misseis aéreos Patriot

**Custo**: 28 soldados mortos e 100 feridos.

**Desastre**: Durante a primeira Guerra do Golfo, um sistema (Patriot) americano de mísseis na Arábia Saudita falhou ao interceptar um míssel vindo do Iraque. O míssel destruiu acampamentos americanos.

**Causa**: Um erro de arredondamento no software calculou incorretamente o tempo, fazendo com que o sistema Patriot ignorasse os mísseis Scud de entrada.

## Pentium Falha em uma Divisão Longa (1993)

**Custo:** $475 milhões e a credibilidade de uma empresa

**Desastre:** O altamente promovido Pentium, da Intel, ocasionalmente cometeu erros ao dividir números de ponto flutuante em um intervalo específico. Por exemplo, dividindo 4195835.0/3145727.0 obteve 1,33374 ao invés de 1,33382, um erro de 0,006%. Embora o bug afetasse apenas alguns usuários, se tornou um pesadelo nas relações públicas. Com uma estimativa de 5 milhões de chips defeituosos em circulação, a Intel se ofereceu para substituir os chips Pentium apenas para os consumidores que poderiam provar que eles precisavam de alta precisão. Contudo a Intel acabou substituindo os chips de qualquer um que reclamou.

**Causa:** O divisor na unidade de ponto flutuante do Pentium tinha uma tabela de divisão falha, faltando cerca de cinco mil entradas, resultando nestes erros de arredondamento.

# Erros absoluto e relativo <a class="anchor" id="chapter_7"></a>

## Erro absoluto <a class="anchor" id="section_7_1"></a>

Sabendo-se que $\pi = (3, 14; 3, 15)$, tomaremos para $π$ um valor $\overline{\pi}$ dentro deste intervalo e teremos, então:

$$\epsilon_{abs_\pi} = |\pi − \overline{\pi}| < 0.01$$

> Seja $\overline{x}$ representado por $\overline{x} = 2112.9$ de forma que $\epsilon_{abs_x}  < 0.1 $ podemos dizer que $\overline{x} ∈ (2112.8, 2113.0)$.


> Seja $\overline{y}$ representado por $\overline{y} = 5.3$ de forma que $\epsilon_{abs_y} < 0.1$, podemos dizer que $\overline{y} ∈ (5.2, 5.4)$
